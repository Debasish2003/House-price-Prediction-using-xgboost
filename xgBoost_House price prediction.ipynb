```python
import xgboost as xgb
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x,y=data.iloc[:,:-1],data.iloc[:,-1]
```

```python
data_dmatrix=xgb.DMatrix(data=x,label=y)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=123)                                 
```

```python
xg_reg=XGBRegressor(objective='reg:linear',colsample_bytrees=0.3,learning_rate=0.1,max_depth=5,alpha=10,n_estimators=10)
```

```python
xg_reg1=XGBRegressor()
```

```python
xg_reg1.fit(xtrain,ytrain)
```

```python
ypred1=xg_reg1.predict(xtest)
```

```python
rmse=np.sqrt(mean_squared_error(ytest,ypred1))
print('RMSE: %.2f'%rmse)
```

## k-fold Cross Validation using XGBoost


```python
params={'objective':'reg:linear','colsample_bytrees':0.3,'learning_rate':0.1,'max_depth':5,'alpha':10,'n_estimators':10}
cv_results=xgb.cv(dtrain=data_dmatrix,params=params,nfold=3,num_boost_round=50,early_stopping_rounds=10,metrics='rmse',as_pandas=True,seed=123)
```

```python
cv_results.head()
```

```python
print((cv_results['test-rmse-mean']).tail(1))
```

```python
xg_reg=xgb.train(params=params,dtrain=data_dmatrix,num_boost_round=10)
```

```python
xgb.plot_importance(xg_reg)
plt.rcParams['figure.figsize']=[5,5]
plt.show()
```
